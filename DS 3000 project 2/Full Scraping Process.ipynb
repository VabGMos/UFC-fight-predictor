```python
import numpy as np
import pandas as pd
import urllib.request as urllib
from bs4 import BeautifulSoup
import time
import random
import re

#initializes data
fight_info = pd.DataFrame()

#event_info
event_names = []
event_dates = []
event_locations = []
#attendance
#event_attendances = []
#list of winners per event
winners = []
#list of fights
fights = []
#list of weight classes per fight
weight_classes = []
#list of methods of win
methods = []
#round of fight end
rounds = []
#time of fight end
times = []


all_events_html = urllib.urlopen('http://ufcstats.com/statistics/events/completed?page=all')
soup = BeautifulSoup(all_events_html.read())
#gets urls of each individual event
event_htmls = [i.get("href") for i in soup.html.body.find_all("a",class_="b-link b-link_style_black")]

print(event_htmls[3])

```

```python
#loops through each link to scrape data
for i in event_htmls:
    print(i)
    fight_html = urllib.urlopen(i)
    soup2 = BeautifulSoup(fight_html.read())
    #event_title
    title = soup2.html.find_all("span",class_="b-content__title-highlight")[0].get_text().strip()
    event_names.append(title)
    
    date_location_attendance = soup2.html.find_all("li",class_="b-list__box-list-item")
    #event date
    date = date_location_attendance[0].get_text().strip().split('\n')[2].strip()
    event_dates.append(date)
    
    #event location
    location = date_location_attendance[1].get_text().strip().split('\n')[3].strip()
    event_locations.append(location)
    
    #event attendance
    #attendance = date_location_attendance[2].get_text().strip().split('\n')[2].strip()
    #event_attendances.append(attendance)
    
    #fight list
    fightersoup = soup2.html.find_all("a", class_= "b-link b-link_style_black")
    fighter_text = [i.get_text().strip() for i in fightersoup]
    fighter1_list = [i for i in fighter_text if fighter_text.index(i) % 2 == 0]
    fighter2_list = [i for i in fighter_text if fighter_text.index(i) % 2 != 0]

    winner_list = fighter1_list
    winners.append(winner_list)

    #list of all fights
    fight_list = []
    for i in range(0, len(fighter1_list)):
        fight_list.append(fighter1_list[i] + ' vs ' + fighter2_list[i])
    fights.append(fight_list)

    #list of all weight classes for fights
    weight_method_soup = soup2.html.find_all("td", class_= "b-fight-details__table-col l-page_align_left")
    #list of weight classes
    weightclass_list = [weight_method_soup[i].get_text().strip() for i in range(1, len(weight_method_soup), 3)]
    weight_classes.append(weightclass_list)
    
    #list of methods of win
    method_list_raw = [weight_method_soup[i].get_text().strip() for i in range(2, len(weight_method_soup), 3)]
    method_list = []
    for i in method_list_raw:
        if '\n' in i:
            i = i[ : i.index('\n')]
        method_list.append(i)
    methods.append(method_list)

    roundsoup = soup2.html.find_all("td", class_= "b-fight-details__table-col")

    #time match was over
    rtime = []
    roundl = []
    for i in range(8, len(roundsoup), 10):
        roundl.append(roundsoup[i].get_text().strip())
    rounds.append(roundl)

    for i in range(9, len(roundsoup), 10):
        rtime.append(roundsoup[i].get_text().strip())
    times.append(rtime)
    
    
```

```python
#Dataframe

fight_info['Title'] = event_names
fight_info['Date'] = event_dates
fight_info['Location'] = event_locations
#fight_info['Attendance'] = event_attendances
fight_info['Fights'] = fights
fight_info['Winners'] = winners
fight_info['Methods'] = methods
fight_info['Rounds'] = rounds
fight_info['Times'] = times

#export_csv = fight_info.to_csv('fight_info_dataframe.csv')

fight_info
```
