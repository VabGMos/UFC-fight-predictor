```python
import numpy as np
import pandas as pd
import urllib.request as urllib
from bs4 import BeautifulSoup
import time
import random
import re

fight_info = pd.DataFrame()

event_name = []
event_date = []
event_location = []
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

html = urllib.urlopen("http://ufcstats.com/statistics/events/completed?page=all")
soup = BeautifulSoup(html.read())
sections = soup.find_all("tr")

for section in sections:
    subsections = section.find_all("td",{"class":"b-statistics__table-col"})
    
    if(len(subsections) == 2):
        event = subsections[0].get_text().strip()
        event_d = None
        for m in months:
            if m in event:
                index = event.find(m)
                event_d = event[index : ]
                event = event[0 : index - 1]
            event = event.strip()
        
        location = subsections[1].get_text().strip()
        event_name.append(event)
        event_date.append(event_d)
        event_location.append(location)

fight_info["Card"] = event_name[1 : ]
fight_info["Date"] = event_date[1 : ]
fight_info["Location"] = event_location[1 : ]

fight_info
    
```
