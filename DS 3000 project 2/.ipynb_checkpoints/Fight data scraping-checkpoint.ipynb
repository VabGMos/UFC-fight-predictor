```python
import numpy as np
import pandas as pd
import urllib.request as urllib
from bs4 import BeautifulSoup
import time
import random
import re

#initializes data
fight_info = pd.DataFrame()

#event_info
event_names = []
event_dates = []
event_locations = []
#attendance
#event_attendances = []
#list of winners per event
winners = []
#list of fights
fights = []
#list of weight classes per fight
weight_classes = []
#list of methods of win
methods = []
#round of fight end
rounds = []
#time of fight end
times = []


all_events_html = urllib.urlopen('http://ufcstats.com/statistics/events/completed?page=all')
soup = BeautifulSoup(all_events_html.read())
#gets urls of each individual event
event_htmls = [i.get("href") for i in soup.html.body.find_all("a",class_="b-link b-link_style_black")]

print(event_htmls[3])

```

```python
#loops through each link to scrape data
for i in event_htmls:
    print(i)
    fight_html = urllib.urlopen(i)
    soup2 = BeautifulSoup(fight_html.read())
    #event_title
    title = soup2.html.find_all("span",class_="b-content__title-highlight")[0].get_text().strip()
    event_names.append(title)
    
    date_location_attendance = soup2.html.find_all("li",class_="b-list__box-list-item")
    #event date
    date = date_location_attendance[0].get_text().strip().split('\n')[2].strip()
    event_dates.append(date)
    
    #event location
    location = date_location_attendance[1].get_text().strip().split('\n')[3].strip()
    event_locations.append(location)
    
    #event attendance
    #attendance = date_location_attendance[2].get_text().strip().split('\n')[2].strip()
    #event_attendances.append(attendance)
    
    #fight list
    fightersoup = soup2.html.find_all("a", class_= "b-link b-link_style_black")
    fighter_text = [i.get_text().strip() for i in fightersoup]
    fighter1_list = [i for i in fighter_text if fighter_text.index(i) % 2 == 0]
    fighter2_list = [i for i in fighter_text if fighter_text.index(i) % 2 != 0]

    winner_list = fighter1_list
    winners.append(winner_list)

    #list of all fights
    fight_list = []
    for i in range(0, len(fighter1_list)):
        fight_list.append(fighter1_list[i] + ' vs ' + fighter2_list[i])
    fights.append(fight_list)

    #list of all weight classes for fights
    weight_method_soup = soup2.html.find_all("td", class_= "b-fight-details__table-col l-page_align_left")
    #list of weight classes
    weightclass_list = [weight_method_soup[i].get_text().strip() for i in range(1, len(weight_method_soup), 3)]
    weight_classes.append(weightclass_list)
    
    #list of methods of win
    method_list_raw = [weight_method_soup[i].get_text().strip() for i in range(2, len(weight_method_soup), 3)]
    method_list = []
    for i in method_list_raw:
        if '\n' in i:
            i = i[ : i.index('\n')]
        method_list.append(i)
    methods.append(method_list)

    roundsoup = soup2.html.find_all("td", class_= "b-fight-details__table-col")

    #time match was over
    rtime = []
    roundl = []
    for i in range(8, len(roundsoup), 10):
        roundl.append(roundsoup[i].get_text().strip())
    rounds.append(roundl)

    for i in range(9, len(roundsoup), 10):
        rtime.append(roundsoup[i].get_text().strip())
    times.append(rtime)
    
    


```

```python
#Dataframe

fight_info['Title'] = event_names
fight_info['Date'] = event_dates
fight_info['Location'] = event_locations
#fight_info['Attendance'] = event_attendances
fight_info['Fights'] = fights
fight_info['Winners'] = winners
fight_info['Methods'] = methods
fight_info['Rounds'] = rounds
fight_info['Times'] = times

#export_csv = fight_info.to_csv('fight_info_dataframe.csv')

fight_info
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score


estimators = {"k-Nearest Neighbor": KNeighborsClassifier(), 
              "Support Vector Machine" : SVC(gamma=0.002), 
              "Gaussian Naive Bayes" : GaussianNB(), 
              "Decision Tree" : DecisionTreeClassifier(),
              }

X_train, X_test, y_train, y_test = train_test_split(features,targ, random_state=3000)
```

```python
#print(type(X_train))
#iterates over estimators
for m in estimators.keys():
    #fits model with data
    
    model = estimators[m].fit(X=X_train, y=y_train)
    print(str(model))
    #predicts accuracy
    accuracy = model.score(X_test, y_test)
    #print statement and formatting
    print(accuracy)

```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

def classifiers_cross_validation():
    kfold = KFold(n_splits=10, random_state=3000, shuffle=True)
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train, y=y_train)
        scores = cross_val_score(estimator=model, X=features, y=targ, cv=kfold)
        #print statement and formatting
        print(m + ":\n         " + f'Mean accuracy: {scores.mean():.2%}' + f'standard deviation={scores.std():.2%}')
classifiers_cross_validation()
```

```python
from sklearn import preprocessing
def preprocessed_regression():
    scaler = preprocessing.MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.fit_transform(X_test)
    #iterates over estimators
    for m in estimators.keys():
        #fits model with data
        model = estimators[m].fit(X=X_train_scaled, y=y_train)
        #predicts accuracy
        accuracy = model.score(X_test_scaled, y_test)
        
        
        #print statement and formatting
        print(m)
        print(accuracy)
        
    return X_train_scaled, X_test_scaled
X_train_scaled, X_test_scaled = preprocessed_regression()
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

def RFE_feature_selection():
    select = RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select = 17)
    
    #fit the RFE selector to the training data
    select.fit(X_train, y_train)

    #transform training and testing sets so only the selected features are retained
    X_train_selected = select.transform(X_train)
    X_test_selected = select.transform(X_test)

    model = GaussianNB().fit(X=X_train_selected, y=y_train)
    
    accuracy = model.score(X_test_selected, y_test)
    
    print("Selected features after RFE:")
    for i in range(0, len(select.get_support())):
        if select.get_support()[i] == True:
            print("\t" + str(features.columns[i]))
    
    print("DecisionTreeClassifier() Regression performance with selected features:")
    print(accuracy)

RFE_feature_selection()
```

```python
def reach_convert(reach):
    if(len(reach)>3):
        return float(reach[0:2])
    else:
        return 0.0
    

maxr = fighter_info["Reach"].map(reach_convert).max()
fighter_info_final["Reach"] = fighter_info_final["Reach"].map(reach_convert)
fighter_info_final["Reach"] = fighter_info_final["Reach"]/maxr
fighter_info_final["Reach"]
```

```python
fighter_info_final.set_index("Name")
reaches = fighter_info_final["Reach"]

def dob_age(dob):
    if(len(dob)>6):
        ind = dob.index(",")
        year = float(dob[ind+1:])
        return (2019-year-15)/45
    else:
        return 0.0
    
def slpm_convert(slpm):
    if(len(str(slpm))==4):
        return slpm/10
    else:
        return 0.0
    
def stracc_convert(stracc):
    if("%" in stracc):
        return float(stracc[0:len(stracc)-1])/100
    else:
        return 0.0
    
    


    
fighter_info_final["Age"] = fighter_info_final["DoB"].map(dob_age)
fighter_info_final["SLpM"] = fighter_info_final["SLpM"].map(slpm_convert)
fighter_info_final["Str Acc"] = fighter_info_final["Str Acc"].map(stracc_convert)
fighter_info_final["SApM"] = fighter_info_final["SApM"].map(slpm_convert)
fighter_info_final["Str Def"] = fighter_info_final["Str Def"].map(stracc_convert)
fighter_info_final["TD Avg"]=fighter_info_final["TD Avg"].map(slpm_convert)
fighter_info_final["TD Acc"]=fighter_info_final["TD Acc"].map(stracc_convert)
fighter_info_final["TD Def"]= fighter_info_final["TD Def"].map(stracc_convert)
fighter_info_final["Sub Avg"]=fighter_info_final['Sub Avg'].map(slpm_convert)
```

```python
fighter_info_final["M-DEC Wins"]=fighter_info_final['Sub Avg']/3
```

```python
fighter_info_final["KO/TKO Wins"]=fighter_info_final["KO/TKO Wins"]/fighter_info_final["KO/TKO Wins"].max()*1.0
fighter_info_final["U-DEC Wins"]=fighter_info_final["U-DEC Wins"]/fighter_info_final["U-DEC Wins"].max()*1.0
fighter_info_final["SUB Wins"]=fighter_info_final["SUB Wins"]/fighter_info_final["SUB Wins"].max()*1.0
fighter_info_final["KO/TKO Losses"]=fighter_info_final["KO/TKO Losses"]/fighter_info_final["KO/TKO Losses"].max()*1.0
fighter_info_final["SUB Losses"]=fighter_info_final["SUB Losses"]/fighter_info_final["SUB Losses"].max()*1.0
fighter_info_final["U-DEC Losses"]=fighter_info_final["U-DEC Losses"]/fighter_info_final["U-DEC Losses"].max()*1.0

```
